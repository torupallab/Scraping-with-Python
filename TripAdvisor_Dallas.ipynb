```python
# This files downloads hotel link, name, overall rating, number of reviews
import requests
from bs4 import BeautifulSoup
import re

out_file = open('hotelsTA_name_id_review.txt', 'w')
def doc_spider(max_pages):
    page = 0
    while page<=max_pages:
        url1='https://www.tripadvisor.com/Hotels-g55711-oa'+ str(page*30) + '-Dallas_Texas-Hotels.html'
        r=requests.get(url1)
        soup=BeautifulSoup(r.content,'lxml')
        title_data = soup.find_all('div',{'class':'listing_info popIndexValidation'}) # It's in list
        iterhotels = iter(title_data)
        next(iterhotels)
        for hotel in iterhotels:
            hotel_name_tag=hotel.find('div',{'class':'listing_title'})
            #print(hotel_name_tag.prettify())
            hotel_name = hotel_name_tag.a.get_text(strip=True) # hotel name
            #print(hotel_name)
            hotel_link= hotel_name_tag.a.get('href') # hotel link
            #print(hotel_link)
            hotel_rating_tag = hotel.find('div',{'class':'rating'}) # hotel overall rating,
            #print(hotel_rating_tag.prettify())
            # some hotels have no rating or review. So,using try and except.
            try:
                hotel_rating_clutter = hotel_rating_tag.img
                hotel_ratingx = str(hotel_rating_clutter).replace('<img alt="','')
                hotel_rating = hotel_ratingx.replace(' of 5 stars" class="sprite-ratings" src="https://static.tacdn.com/img2/x.gif"/>','')
                num_of_review = hotel_rating_tag.a.get_text(strip=True).replace(' Reviews','') # total number of review
            except AttributeError:
                hotel_rating = 'NA'
                num_of_review = 'NA'
            
            out_file.write('\t'.join([hotel_link, hotel_name, hotel_rating, num_of_review]) + '\n')
        page+=1


doc_spider(6)
out_file.close()   
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen

out_file = open('hotel_review1.txt', 'w', encoding='utf-8')

with open ('hotelsTA_name_id_review_input.txt','r') as input_file:
    line = input_file.readline()
    while line != '':
        entries = line.split('\t')
        hotel_link = entries[0]
        #print(hotel_link)
        #url2='https://www.tripadvisor.com'+str(/Hotel_Review-g55711-d122703-Reviews-Grand_Hyatt_DFW-Dallas_Texas.html)+'#REVIEWS'
        url2='https://www.tripadvisor.com'+str(hotel_link)+'#REVIEWS'
        html = urlopen(url2)
        page_content = html.read().decode('utf-8','ignore') 
        soup=BeautifulSoup(page_content,'lxml')
        address_tag = soup.find('div',{'property':'address'})
        address = address_tag.get_text(strip= True)
        #print(address)
        total_page_tag = soup.find('div',{'class':'pageNumbers'})
        #print(total_page_tag.prettify())
        #print((total_page_tag.find_all('a'))[-1].get_text()) 
        try:
            total_page_number = (total_page_tag.find_all('a'))[-1].get_text() # getting the total numbe of pages
        except AttributeError:
            total_page_number = '1'
        #print(total_page_number)
        out_file.write('\t'.join([hotel_link, total_page_number, address]) + '\n')
        
        line = input_file.readline()
input_file.close()
out_file.close()
```
